```python
import pandas as pd
import os

# read folder ../memories and compile a list of all gif, jpg and png
def get_image_files(folder):
    image_files = []
    for root, dirs, files in os.walk(folder):
        for file in files:
            if file.lower().endswith(('.gif', '.jpg', '.jpeg', '.png', '.mp4', '.mov')):
                image_files.append(os.path.join(root, file))
    return image_files

images = get_image_files('../memories')
```

```python
for i, image in enumerate(images):
    #convert mp4 into gif
    if image.lower().endswith(('.mp4', '.mov')):
        # Use ffmpeg to convert mp4 to gif
        gif_name = image[:-4] + '.gif'
        os.system(f'ffmpeg -i "{image}" "{gif_name}"')
        images[i] = gif_name  # Update the list to use the new gif file
    else:
        # Ensure the image is a gif, jpg, or png
        if not image.lower().endswith(('.gif', '.jpg', '.jpeg', '.png')):
            continue
    
```

```python
from PIL import Image
import shutil

# Create the 'resized' folder if it doesn't exist
resized_folder = 'resized'
os.makedirs(resized_folder, exist_ok=True)

for i, image in enumerate(images):
    ext = os.path.splitext(image)[1].lower()
    resized_path = os.path.join(resized_folder, os.path.basename(image))
    
    # For images (jpg, jpeg, png)
    if ext in ('.jpg', '.jpeg', '.png'):
        with Image.open(image) as img:
            # Calculate the new size while keeping the aspect ratio
            aspect_ratio = img.width / img.height
            if aspect_ratio > 1:  # Landscape
                new_width = 250
                new_height = int(new_width / aspect_ratio)
            else:  # Portrait or square
                new_height = 250
                new_width = int(new_height * aspect_ratio)
            
            img = img.resize((new_width, new_height), Image.LANCZOS)
            img.save(resized_path)
        print(f"Resized {image} to {new_width}x{new_height} and saved to {resized_path}.")
    
    # For GIFs
    elif ext == '.gif':
        os.system(f'ffmpeg -i "{image}" -vf "scale=250:-2,fps=10" "{resized_path}"')
        print(f"Resized {image} to 250x250, 10fps, and saved to {resized_path}.")
    
    # Unsupported formats
    else:
        print(f"Skipped {image} (unsupported format).")

```

```python
ids = []
imagess = []
for image in images:
    print(image)
    ids.append(os.path.splitext(os.path.basename(image))[0])
    imagess.append(os.path.abspath(image))
df = pd.DataFrame({'title': ids, 'url': imagess})
df.to_csv('data_new.csv', index=False)
    
```
